```python
df = pd.read_csv('Data.csv')
```

```python
min_lat = 37.77820
max_lat = 37.77923
min_lon = -122.44587
max_lon =-122.44439
```

```python
number_of_users = 535

for i in range(1,10):
    for j in range(150):
        user_id = np.random.randint(0,number_of_users)
        trajectory = df.loc[df["users"] == user_id]
        plt.scatter(trajectory["lon"], trajectory["lat"], color='black', s=20)
    plt.ylim(min_lat,max_lat)
    plt.xlim(min_lon,max_lon)
    plt.axis("off")
    plt.savefig("/home/hussein/Hola/US/im/{}.jpg".format(i+random.randrange(1000000000)))
```
